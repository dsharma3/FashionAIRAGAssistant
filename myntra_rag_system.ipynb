# Embeddings

```python

import pandas as pd
import numpy as np

df = pd.read_csv("Fashion Dataset v2.csv")
df["ratingCount"].fillna(0, inplace=True)
df["avg_rating"].fillna(df["avg_rating"].mean(), inplace=True)

def build_text(row):
    return f"""
    Name: {row['name']}
    Brand: {row['brand']}
    Colour: {row['colour']}
    Product: {row['products']}
    Price: {row['price']}
    Description: {row['description']}
    Attributes: {row['p_attributes']}
    Avg Rating: {row['avg_rating']} (from {int(row['ratingCount'])} ratings)
    """
df['document'] = df.apply(build_text, axis=1)
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
```

```python
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.config import Settings
from chromadb.errors import NotFoundError

# Initialize model and client
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
import chromadb

chroma_client = chromadb.PersistentClient(path="./chromadb_data")
collection = chroma_client.get_or_create_collection("fashion_db")

# Prepare and clean data
df_clean = df[['p_id', 'document']].dropna()
df_clean['p_id'] = df_clean['p_id'].astype(str)

# Helper: Get existing IDs in batches
def get_existing_ids(candidate_ids, batch_size=5000):
    existing = set()
    for i in range(0, len(candidate_ids), batch_size):
        batch_ids = candidate_ids[i:i+batch_size]
        try:
            res = collection.get(ids=batch_ids)
            existing.update(res['ids'])
        except Exception as e:
            print(f"Skipping batch due to error: {e}")
    return existing

candidate_ids = df_clean['p_id'].tolist()
existing_ids = get_existing_ids(candidate_ids)
print(f"{len(existing_ids)} documents already exist in collection.")

# Filter only new items
new_df = df_clean[~df_clean['p_id'].isin(existing_ids)]

# Helper: Add in batches to avoid hitting ChromaDB limits
def add_documents_in_batches(collection, df_to_add, batch_size=5000):
    if df_to_add.empty:
        print("No new documents to add.")
        return

    docs = df_to_add['document'].tolist()
    ids = df_to_add['p_id'].tolist()

    for i in range(0, len(docs), batch_size):
        batch_docs = docs[i:i+batch_size]
        batch_ids = ids[i:i+batch_size]
        batch_embeddings = model.encode(batch_docs, show_progress_bar=True)
        collection.add(documents=batch_docs, embeddings=batch_embeddings, ids=batch_ids)
        print(f"Added batch of {len(batch_docs)} documents.")

# Add new docs
add_documents_in_batches(collection, new_df)
# all persist() to save changes
```
